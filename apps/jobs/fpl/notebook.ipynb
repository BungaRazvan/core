```python
import requests
import pandas as pd
import numpy as np

pd.options.display.max_rows = 999
```

```python
json = requests.get('https://fantasy.premierleague.com/api/bootstrap-static/').json()
```

```python
elements_df = pd.DataFrame(data=json['elements'], columns=['first_name','second_name','team','element_type','selected_by_percent','now_cost','minutes','transfers_in','value_season','total_points'])
elements_types_df = pd.DataFrame(json['element_types'])
teams_df = pd.DataFrame(data=json['teams'],)
```

```python
elements_df
```

```python
elements_df['position'] = elements_df.element_type.map(elements_types_df.set_index('id').singular_name)
```

```python
elements_df['team_name'] = elements_df.team.map(teams_df.set_index('id').name)
```

```python
elements_df['roi'] = elements_df['value_season'].astype(float)
```

```python
elements_df.sort_values(['roi', 'minutes'], ascending=False)
```

```python
players_df = elements_df.filter(['first_name', 'second_name', 'minutes', 'roi', 'now_cost', 'total_points', 'position', 'team_name'], axis=1)
players_df.sort_values(['minutes',  'roi'], ascending=False)
```
